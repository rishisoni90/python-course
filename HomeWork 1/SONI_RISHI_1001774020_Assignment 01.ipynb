   - if it user gave any float value then it going to be checked in the if loop i.e ( . ) and going to be stored in a variable.
   - if it user gave integer value then it going to check and stored in a variable.
3. Thus for prime , composite and other statement , we will use our condition statements.
   - if variable is 1 or 0 then it's neither prime nor composite.
   - if variable is < 0  print "You did not enter positive number" statement.
   - if variable is divisible by 1 and variable by itself then it's prime otherwise composite.

```python
#  HomeWork 1 Problem 3


# Thus we have 6 output So we are doinging in that format ....

for x in range(6): # Number of times , we asking the input from user.....
    
    #... Taking the input in the form of String ... Because we are not sure what user gonna type , it may be Integer or float or String...
    stringnumber = input("Enter positive number: ")
    # .. Converting String input into a number ...
    
    if stringnumber.isdigit():
        # .. Checking if the input is a integer or a float value ....
        if("."in stringnumber):
            number = float(stringnumber)
        else:
            number = int(stringnumber)
            
            # Now putting conditions ... If the values < 0 then the we print our output...
        if(number<0):
            print("You did not enter positive number")
        elif(number == 1 or number ==0):
            #... If number is 1 or 0 then priting our output...
            print("It is neither prime nor composite")
            
            # Checking condition and again printing our output............
        elif(type(number) == int):

            for i in range(2,number):
                if number % i == 0: # Checking the prime condition , if its true then its not prime...
                    print("It is a composite ")
                    break
            else:
                print("It is a prime")
               
        else:
            print("You did not enter a positive integer")
    else:
        print("You did not enter positive number")
        
# PS : 
# PYTHON DOCUMENTATION FOR REFERENCE FOR SYNTAX AND OTHER INBUILT FUNCTIONALITY....
# INSY 5336 LECTURES AND MODULES..
```
