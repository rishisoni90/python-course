# HOMEWORK 2 QUESTION 1 :

Step1: We created a function named positivityGenerator. The Function takes the parameter String S.
Step2: Then converting the given String into Upper Case and save it to another variable named X.
Step3: Finding the given words in the String X, find the word 'NOT' and save it to another variable N.
Step4: Find the word 'BAD' and save it to another vairable i.e B.
Step5: Compare the ordering of the word. If 'not' appears before the 'bad', the function should replace the whole 'not'...'bad' substring with 'good'.
STEP6: Output is generated by calling the function.

```python
def positivityGenerator(s):
   
    X = s.upper() # Converting String to UPPER CASE..

    N = X.find("NOT") # FINDING THE WORD NOT IN STRING ..

    B = X.find("BAD") # FINDING THE WORD BAD IN STRING..
 
    if N < B:  # COMPARING THE APPEARANCE OF THE CHARACTER AND GENERATING OUTPUT ACCORDINGLY.

        answer = s[:N] + "good" + s[(B+3):]

        return answer

    else:

        return s


s = input("Input = ")

print("Output = ",positivityGenerator(s)) # CALLING THE FUNCTION AND GETTING THE OUTPUT.
```

# HOMEWORK2 QUESTION 3:

STEP1: Taking two Strings inputs from the user. First is the string and the second one is choice.
STEP2: If the choice entered is odd, then it will go to odd loop otherwise it will go to elif loop.
STEP3: In IF loop, using list comprehension to extract list of elements at only odd places using join to convert list to string.
STEP4: In Elif loop, using list comprehension to extract list of elements at only even places
using join to convert list to string.
STEP5: In Else loop, using list comprehension to extract list of elements at only nth places
using join to convert list to string.
STEP6: Print output.

```python
# prompting user for input
string = input("Input string = ")
# prompting user for choice
choice = input("Input choice = ")
# processing according to choice
if choice =="odd":
    # using list comprehension to extract list of elements at only odd places
    # using join to convert list to string
    output = "".join([e for i,e in enumerate(string) if i%2])
elif choice == "even":
    # using list comprehension to extract list of elements at only even places
    # using join to convert list to string
    output = "".join([e for i,e in enumerate(string) if (i+1)%2])
else:
    # using list comprehension to extract list of elements at only nth places
    # using join to convert list to string
    output = "".join([e for i,e in enumerate(string) if (i+1)%int(choice)])

# display result
print("Output =",output)
```

```python
num = int(input("How many elements? "))

list1 = [] # LIST 1 IS CREATED.
list2 = [] # LIST 2 IS CREATED.

for i in range(1,num+1) :
    list1.append((input("Enter the "+str(i)+"th value of List 1 : ")))
for i in range(1,num+1) :
    list2.append((input("Enter the "+str(i)+"th value of List 2 : ")))

for i in range(num) :
    for j in range(num) :
        print(str(list1[i])+str(list2[j])) # OUTPUT IS GENERATED..
```
