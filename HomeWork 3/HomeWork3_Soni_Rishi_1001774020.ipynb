# HOMEWORK 3 QUESTION 1:

STEP1: GETTING THE FILE NAME FROM THE USER.
STEP2: WE CREATED 4 VARIABLES WHERE WE STORE OUR TOTAL NUMBER OF LINES, NUMBER OF WORDS, NUMBER OF CHARACTER AND WORD LENGTH.
STEP3: WE OPEN THE FILE WITH READ r and STORE THAT FILE INTO THE VARIABLE NAMED, fhand.
STEP4: TRAVERSING THE FILE BY USING THE FOR LOOP.
STEP5: WE USING THE SPILT METHOD TO FIND THE NUMBER OF LINES, NUMBER OF WORDS CAN BE CALCULATED BY ADDING THE NUMBER OF WORDS OF THE CURRENT LINE. TOTAL NUMBER OF CHARACTER CAN BE ADDED TO THE CURRENT LINE TO GET TOTAL CHARACTER. 
BY DIVING CHARACTER AND TOTAL WORDS, WE CAN GET THE WORD LENGTH.
STEP6: PRINT OUR DESIRE OUTPUT.

```python
#Get filename from user
fileName=input("Enter the filename: ")
numLines=0
numWords=0
numChars=0
wordLen=0

try:
   #open file
   fhand=open(fileName,"r")
   #traverse the file by lines
   for line in fhand:

       numLines=numLines+1
       #Turn the current line into a list of words
       lineList=line.split()

       #Calculate the total number of words by adding the number of words of the current line
       numWords=numWords+len(lineList)

       #lineStr is a string created by concatenating all the words of the current line while the spaces between the words are removed
       lineStr="".join(lineList)

       #Calculate the total number of characters by adding the number of characers of the current line(excluding spaces)
       numChars=numChars+len(lineStr)

   fhand.close()
   wordLen=(numChars/numWords)

   print("Number of line:",numLines)
   print("Number of words:",numWords)
   print("Number of characters:",numChars)
   print("Average length of a word:",wordLen)

except:
   print("File not found!")
```

# HOMEWORK 3 QUESTION 2:

STEP1: PASSING FILENAME TO THE FUNCTION read_file(filename).
STEP2: OPEN THE FILE AND READING THE FILE WITH readlines(). Storing it to Contents.
STEP3: LOOPING EACH CONTENT OF THE FILE BY USING FOR LOOP.
STEP4: I USED STRIP AND SPLIT TO REMOVE LEADING AND TRAILING WHITESPACES AND SPLIT THE STRING.
STEP5: GET THE REVIEWER NAME FROM THE DATASET.
STEP6: CREATING EMPYT DICTIONARY AND SETTIGN THE ABOVE VALUE IN THE DICTIONARY.
STEP7: WE ADD TWO FIELDS IN DICTIONARY , 1 IS REVIEWER NAME AND SECOND ONE IS MOVIE RATING.
STEP8: NEW FUNCTION FOR THE REVIEWER 1 AND REVIEWER 2.NOW TO CHECK THE REVIEW IS PRESENT IN THE DATABASE OR NOT, BY USING THE IF STATEMENT.
STEP9: IF IT CONTAIN THEN WE CALCULATE THE SCORE AND RETURE IT.
STEP10:NEW FUNCTION FOR THE REVIEWER 1 AND ALL THE REVIEWER IN THE DATABASE.NOW TO CHECK THE REVIEW IS PRESENT IN THE DATABASE OR NOT, BY USING THE IF STATEMENT.
STEP11: IF THE REVIEWER IS PRESENT, THEN FINDING THE SUM OF SQUARE OF DIFFERENCE BETWEEN THE RATING OF COMMON MOVIE BETWEEN REVIEWER 1 AND REST OF THE REVIEWER.
STEP12: MAIN FUNCTION IS TO PROVID ALL THE USER INPUT TO THE ABOVE FUNCTION.
STEP13: DISPLAY DESIRE OUPUT.

```python
# AUTHOR RISHI KUMAR SONI, UTA ID 1001774020.
def read_file(filename):
   ''' Function to read an input file and return a dictionary where
   keys are names of reviewers and values are nested dictionary whose keys are the
   movie names and values are the ratings for that movie by the the reviewer
   '''
  
   # create an empty output dictionary
   movies_db = {}
   # open th file
   with open(filename) as fp:
       contents = fp.readlines() # read the contents of the file into the list with each line being an element of the list
      
       # loop over each element of contents
       for line in contents:
           # strip is used to remove leading and trailing whitespace from line
          
           # split is used to split the string using the provided delimiter
           
           # Here it splits the line using comma as the delimiter and returns a list of strings  
           data = line.strip().split(",")
           # get the reviewer name which is the first element of the data list
           reviewer_name = data[0]
           # set the value for the reviewer in the dictionary to empty dictionary
           movies_db[reviewer_name] = {}
           i = 1
           # loop over the data list starting from index 1
           while(i < len(data)):
               # populate the dictionary for the reviewer where ith index contains the movie name
               # and (i+1)th index contains the rating for that movie by the reviewer
               movies_db[reviewer_name][data[i]] = float(data[i+1])
               i += 2 # ignore the rating since it has been considered in the previous step
      
       fp.close() # close the file
      
   return movies_db  
  
  
def similarity_between_2_reviewers(movies_db, reviewer1, reviewer2):
   ''' Function to calculate and return the similarity score between reviewer1 and reviewer2
   '''
  
   # check that both reviewer1 and reviewer2 are in the database
   if reviewer1 in movies_db and reviewer2 in movies_db:
       # get the dictionary of movies and ratings for reviewer1 and reviewer2
       movies_ratings1 = movies_db[reviewer1]
       movies_ratings2 = movies_db[reviewer2]
      
       sum_sq = 0
       # loop to caluclate the sum of square of differences between the ratings of the common movies
       # between reviewer1 and reviewer2
       for movie in movies_ratings1:
          
           if movie in movies_ratings2:
               sum_sq += (movies_ratings1[movie] - movies_ratings2[movie])**2
              
              
       return (sum_sq**0.5) # calculate and return the similarity score
   return None # either reviewer1 or/and reviewer2 not in database


def similarity_all_reviewers(movies_db, reviewer):
   ''' Function to calculate and return the similarity scores between the reviewer and
       all other reviewers in the database
   '''  
   # check that reviewer is present in the database
   if reviewer in movies_db:
       reviewer_rating = {} # create an empty output dictionary
      
       # get the dictionary of movies and ratings for reviewer
       movies_ratings = movies_db[reviewer]
      
       # loop over the database
       for reviewer1 in movies_db:
           # if reviewer is not same as reviewer1
           if reviewer != reviewer1:
               sum_sq = 0
               # get the dictionary of movies and ratings for reviewer1
               movies_ratings1 = movies_db[reviewer1]
              
               # loop to caluclate the sum of square of differences between the ratings of the common movies
               # between reviewer1 and reviewer
               for movie in movies_ratings:
                  
                   if movie in movies_ratings1:
                       sum_sq += (movies_ratings[movie] - movies_ratings1[movie])**2
                      
               reviewer_rating[reviewer1] = sum_sq**0.5 # set the dictionary element for reviewer1
       return reviewer_rating
   return None # reviewer not present in dictionary

# main function
def main():
  
   choice = 0
   # input of filename
   filename = input('Give the name of the movie reviews file: ')
   movies_db = read_file(filename) # get the dictionary representing the database
   # check if file contains records
   if len(movies_db) == 0:
       print('No reviews found')
   else:  
       # loop that continues till the user wants
       while choice != 3:
           # input of choice
           choice = int(input('\nWhat do you want to do? Input 1 for similarity between two reviewers, or Input 2 for similarity between one reviewer and all others in the database or 3 to quit: '))
          
           # perform the operation based on the choice
           if choice == 1:
               reviewer1 = input('Provide Reviewer1 name: ')
               reviewer2 = input('Provide Reviewer2 name: ')
              
               score = similarity_between_2_reviewers(movies_db, reviewer1, reviewer2)
              
               if score == None:
                   print('\nOne or more reviewer(s) not found in the database')
               else:
                   print('\nThe similarity score between Peter and Trevor Chappell is: %.3f' %score)
                  
           elif choice == 2:
               reviewer = input('Provide Reviewer name: ')
               score_dic = similarity_all_reviewers(movies_db, reviewer)
              
               if score_dic == None:
                   print('\nReviewer not found in the database')
               else:
                   print('\nThe Similarity Scores are:')
                   for reviewer1 in score_dic:
                       print('%s %s %.3f' %(reviewer, reviewer1, score_dic[reviewer1]))
          
           elif choice != 3:
               print('Invalid choice')
              
       print('Goodbye!')
      
#call the main function
main()      
```
